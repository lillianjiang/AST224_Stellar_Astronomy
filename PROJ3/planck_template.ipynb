# units!

# import packages

```python
import numpy as np                       #numpy enables functions for data arrays
import matplotlib.pyplot as plt          #matplotlib enables plotting routines
```

# constants in cgs units

```python
#The Planck Constant
h = 6.626 * 10**(-27)  #erg*sec

#The Boltzmann Constant
k = 1.381 * 10**(-16)  #erg/K

#Speed of light
c = 2.998 * 10**(10)   #cm/sec
```

# enter temp, lambda

```python
temp = np.arange(1000, 10000, 1000)         
#in nm unit
lambdaRange_nm = np.arange(100,5000,100)
#convert into cm
lambdaRange_cm = lambdaRange_nm*(1e-7)
```

```python
# meshgrid command will create two new 2-d arrays from temp and wave
# these 2d arrays will be used to calculate B_dummy element by element

Temp, WaveRange = np.meshgrid(temp,lambdaRange_cm)        # create new arrays Temp, Wave
print("shape Temp =", np.shape(Temp)), print("Temp =")
print(Temp)     
print("shape WaveRange =",np.shape(WaveRange)), print("WaveRange = ") 
print(WaveRange)  
```

# calculate Planck

```python
calc_top = 2*h*c**2
expo = (h*c)/(WaveRange*k*Temp)
calc_bot = (WaveRange**5)*(np.exp(expo)-1)
planck_result = calc_top/calc_bot
```

# plot Planck

```python
plt.figure(figsize=(20,30))
plt.subplot(3,1,1)
plt.plot(lambdaRange_nm,planck_result,'*-')
plt.title('The Planck function for Temp = 888 K ',size=20)
plt.ylabel('Intensity',size=20)
plt.xlabel('wavelength in nm',size=20)

plt.subplot(3,1,2)
plt.plot(lambdaRange_nm,np.log10(planck_result),'*-')
plt.title('The Planck function for Temp = 888 K With log 10 of Intensity',size=20)
plt.ylabel('log 10 of Intensity',size=20)
plt.xlabel('wavelength in nm',size=20)

plt.subplot(3,1,3)
plt.plot(np.log10(lambdaRange_nm),np.log10(planck_result),'*-')
plt.title('The Planck function for Temp = 888 K With log 10 of Intensity and Wavelength',size=20)
plt.ylabel('log 10 of Intensity',size=20)
plt.xlabel('log 10 of wavelength in nm',size=20)
```

```python
# plot Planck function as a function of wave for each temp
plt.figure(figsize=(10,10))
plt.plot(lambdaRange_nm, planck_result,'-', markersize = 15)
plt.ylabel("Intensity ", size=15)
plt.xlabel("wavelength(nm)",size=15)
plt.title('Planck Function from meshgrid', size = 25)
plt.grid()

# add legend with temp values
labels = (temp)
plt.legend(labels)
# if there are a lot of temps this can get messy - see next cell for alternate approach
```
